Ploting data in python is very similar to the process you do in linear algebra: Given a sample of X values you use a function to obtain the Y values an plot the intersection of the X and Y coordinates in the carthesian plane. When we don't have real data to plot, we can use numpy to generate some sample values, for example:

```python
import numpy as np
import matplotlib.pyplot as plt

x_axis = np.linspace(-3,3,20)  # by default linspace displays 50 values
x_axis
```

We can use these values as "seed" or x parameters to obtain the Y values using any linear algebra function. This is an example using the square function.

```python
x = np.linspace(-3,3,20)

def f(x):
    return x**2

y = f(x)

plt.figure(figsize=(5,3))
plt.plot(x,f(x))
```

#### The Parent functions in linear algebra
Data patterns can be described by algebra functions and you can plot and visualize them in charts. 
The following are called the "parent functions" in linear algebra and the will usually cover 95% of your data patterns.

#### Square Function

```python
x = np.linspace(-3,3,1000)
def square(x):
    return x**2

plt.figure(figsize=(5,3))
plt.plot(x, square(x))
```

#### Sen function 

```python
def sen_function(x):
    return np.sin(x)

plt.figure(figsize=(5,3))
plt.plot(x, sen_function(x))
```

#### Exponential function

```python
def exponential(x):
    return 100*(np.power(2, x))

plt.figure(figsize=(5,3))
plt.plot(x, exponential(x))
```

#### Logarithmic function

```python
def logarithmic(x):
    return np.log(x)

plt.figure(figsize=(5,3))
plt.plot(x, logarithmic(x))
```
